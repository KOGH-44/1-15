### P-value 와 T-test, T-statistics

```python
import numpy as np # 넘파이 임포트
import pandas as pd # 판다스 임포트
from scipy import stats # 사이파이런 임포트

# 두 그룹의 데이터 생성 (예: A와 B 그룹)
np.random.seed(100)  # 결과 재현을 위한 시드 설정
group_a = np.random.normal(loc=30, scale=10, size=100)  # 평균 30, 표준편차 10 (임의 지정)
group_b = np.random.normal(loc=35, scale=10, size=100)  # 평균 35, 표준편차 10 (임의 지정)

# 데이터프레임 생성
df = pd.DataFrame({'Group': ['A'] * 100 + ['B'] * 100, 'Value': np.concatenate([group_a, group_b])})

# 두 그룹 간의 평균 차이 검정 (t-검정)
t_statistic, p_value = stats.ttest_ind(group_a, group_b) # 두개의 독립된 귀무가설이 있다는 것을 검증하기 위한 메서드 ttest_ind()를 활용하여 그룹 a와 그룹 b를 t-검정(t-test)한다. 
# 일련의 과정을 거쳐서 t 통계량과 prob 값을 리턴한다.

# 결과 출력
print("t-통계량:", t_statistic) 
# t 통계량이 0에 가까운 값이면 그룹 a,b 간 평균 차이가 통계적으로 의미가 없음. 양수의 경우는 표본 평균이 a가 더 크고, 음수의 경우는 표본 평균이 b가 더 크다.
print("p-값:", p_value)
# p 값이 0.05 이하일 경우 귀무가설이 통계적으로 무의미하며, 대립가설이 통계적으로 유의미하다는 걸 의미하고 대립가설을 채택할 수 있게 된다.
# p 값이 0.05 초과일 경우 귀무가설이 통계적으로 유의미하며, 대립가설이 통계적으로 무의미하다는 걸 의미하고 귀무가설을 채택할 수 있게 된다. 
# 위의 경우, 귀무가설 : 그룹 a 와 그룹 b의 평균은 같다.    대립가설 : 그룹 a 와 그룹 b의 평균은 다르다.


# 데이터프레임 확인
print(df.head())
```

위의 결과의 경우 t 통계량은 -3.690...이므로 A 그룹의 표본 평균이 B 그룹의 표본 평균보다 작다는걸 알 수 있으며, 
p 값은 0.000289...< 0.05 이므로 두 그룹의 평균은 같다는 귀무가설은 통계적으로 무의미하므로, 두 그룹의 평균은 다르다는 대립가설이 통계적으로 유의미함을 의미한다.
따라서 두 그룹의 평균은 다르다고 해석 할 수 있다.

※이 예제에서 귀무가설이 "두 그룹의 평균은 같다" 가 된 이유: 귀무가설(H0)은 일반적으로 '효과가 없다' 또는 '차이가 없다'는 가설이므로 평균에 관한 t-검정에서 귀무가설은 "두 그룹의 평균은 같다"의 형태를 가지게 된다. (연구자들이 이러한 귀무가설을 세우는 이유는 두 그룹간의 차이가 우연에 의한 것이 아니라는 증거를 찾아내기 위한 대립가설의 타당성을 내세우기 위한 것)

※ t-통계량을 구하는 이유: t-통계량은 p-값을 계산하는 중간단계이다. 이러한 t-통계량은 p-값을 구하는데 사용됨. 또, t-통계량은 그룹 간의 평균 차이를 표준 편차로 나눈 값이므로, 효과의 크기를 나타낸다. 또, t-통계량을 통해 표준화된 비교를 할 수 있다. 연구나 실험 결과를 통일된 척도로 비교 할 수 있다.

 

### KMeans 

```python
import numpy as np # 넘파이를 np라는 이름으로 임포트
import matplotlib.pyplot as plt # matplotlib 파일의 pyplot 함수를 plt란 이름으로 임포트
from sklearn.cluster import KMeans # sklearn 파일의 cluster 클래스 안에 있는 KMeans 메서드 임포트
from sklearn.datasets import make_blobs # sklearn 파일의 datasets 클래스 안에 있는 make_blobs 메서드 임포트

# 가상의 데이터 생성
X, _ = make_blobs(n_samples=300, centers=4, random_state=42) # 4개의 클러스터를 만들도록 하는 300개의 샘플을 가진 가상 데이터의 샘플 생성 (random_state = 42 <= 다른 사람이 돌렸을때도 동일한 데이터 나오게 제어하는 갑)
# 이때 X 변수는 Numpy Array 자료구조이다.

# Elbow Method를 통한 최적의 K 찾기
inertia_values = []  # 클러스터링 비용을 저장할 리스트 초기화

# K를 1부터 10까지 변경하면서 K-Means 모델 훈련 및 클러스터링 비용 저장
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X) # 1부터 11까지 바꾸면서 학습
    inertia_values.append(kmeans.inertia_) # 클러스터 내의 제곱거리의 합을 구하여 inertia_values 리스트에 추가 // kmeans.fit(X)함수 이후 X 배열의 inertia_ 값이 k에 맞추어 갱신
    

# Elbow 그래프 시각화
plt.plot(range(1, 11), inertia_values, marker='o') # elbow method를 통해 최적의 K값을 찾아낸다. K 이후로 inertia가 급격하게 줄지 않는 곳을 선택하면 좋다. Elbow 이외에도 silhouette score, Davies-Bouldin Index과 같은 방법이 존재
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Cost)')
plt.show()

# K-Means 모델 생성
kmeans = KMeans(n_clusters=4, random_state=42) # 클러스터의 개수를 3개로 설정하는 객체 생성 완료
kmeans.fit(X) # 위에서 만든 샘플 데이터 X를 클러스터링한다.  fit() 메서드는 모델을 주어진 데이터에 적합하게 만들도록 학습시킨다.
# X에는 특성이나 레이블을 포함한 모든 정보를 담고 있다.
# 위 작업 이후 kmeans 객체에는 cluster_centers_나 labels_같은 값이 채워지게 된다.

# 클러스터 중심 확인
centroids = kmeans.cluster_centers_
print("클러스터 중심:\n", centroids) # 클러스터의 각각의 좌표를 알 수 있음

# 클러스터 할당 확인
labels = kmeans.labels_
print("클러스터 할당:\n", labels) #각각의 샘플들이 어떠한 클러스터에 할당됬는지 알수 있음

# 데이터와 클러스터 중심 시각화 (산점도)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k', alpha=0.4) # X[:, 0]은 X의 모든 행의 0열 값이므로 X의 x좌표, X[:, 1]은 X의 모든 행의 1열 값이므로 X의 y좌표,
# c = labels 는 라벨링 구분을 색상으로 하겠다는 뜻, cmap='viridis'는 컬러맵이 연속적인 색 증가를 가짐, alpha 값은 데이터포인트들이 1에 가까울수록 투명하지 않아지고, 0에 가까울수록 투명해진다. 
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids') 
plt.title('K-Means Clustering') # 표 이름을 K-Means Clustering으로 정함 
plt.xlabel('Feature 1') # x축 라벨링을 Feature 1 으로 함
plt.ylabel('Feature 2') # y축 라벨링을 Feature 2 으로 함
plt.legend() # 범례(legend) 자동으로 추가   
plt.show() # 도표 표시
X
```


- elbow method에서 inertia(cost) 값의 하락이 급격히 줄어드는 첫 K값을 선택하는 이유는 아무래도 굳이 많은 K값을 통한 K개 이상의 군집으로의 군집화가 무의미해서 일 것이라 생각한다.
